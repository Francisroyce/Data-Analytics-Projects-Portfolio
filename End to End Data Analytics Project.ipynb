```python
#let pass unknown vlaues to be treated as null
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#fixing column names. make them lowercase and replace space with underscore
#df.rename(columns={'Order Id':'order_id', 'City':'city'})
df.columns = df.columns.str.lower()
df.columns=df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df.head(10)
```

```python
 # drive new columns discount, sale price and profit
df['list_price']*df['discount_percent']*.01
df['discount'] = df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
#checking data types
df.dtypes
```

```python
#convert order date from object to datetime
df['order_date']= pd.to_datetime(df['order_date'], format='%Y-%m-%d')
df
```

```python
#dropping collumns we are not suing anymore(cost price, list price, and discount percent)
df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace=True)
```
