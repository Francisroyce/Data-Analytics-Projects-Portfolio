 # Understanding the Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('payment_report (3).csv')
df
```

# Check for Missing Values

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

# Remove Unnecessary Columns

```python
# Drop unnecessary columns
columns_to_drop = ['Invoice Number']  
df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
df
```

# Renaming of the 'Unknown column to serial number'

```python
df = df.rename(columns={'Unnamed: 0': 'Serial_Number'})
```

```python
df
```

```python
#fixing column names. make them lowercase and replace space with underscore
df.columns = df.columns.str.lower()
df.columns=df.columns.str.replace(' ', '_')
```

```python
df
```

# Convert Date Columns to Datetime Format

```python
df['payment_date'] = pd.to_datetime(df['payment_date'], format='%d/%m/%Y', dayfirst=True)
df
```

```python
#checking datatypes
df.dtypes
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

# Fill Missing Categorical Values

```python
# Fill missing categorical values
categorical_columns = ['industry', 'payment_method', 'payment_bank']
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
```

```python
df.isnull().sum()
```

# Handle Duplicate Records

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
df.shape
```

# Save the Cleaned Dataset

```python
df.to_csv("cleaned_data.csv", index=False)
```
